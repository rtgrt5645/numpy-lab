## Revenue Insights

```python
# 1. Total Monthly revenue
monthly_sales = kababjees_sales_analysis[: , 1:]
monthly_revenue = monthly_sales.sum(axis=1)

for month in range(len(monthly_revenue)):
    print(f"{month+1} Month Revenue: {monthly_revenue[month]}")
```

```python
plt.plot(kababjees_sales_analysis[:, 0], monthly_revenue, marker='o', color='teal')
plt.title("Monthly Total Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue (PKR)")
plt.grid(True)
plt.show()
```

```python
# 2. Revenue growth rate (%)
monthly_growth_rate = (monthly_revenue[1:] - monthly_revenue[:-1])/monthly_revenue[:-1]*100
for i, rate in enumerate(monthly_growth_rate, start=2):
    print(f"Month {i} Growth Rate(%): {rate:.2f}%")
```

```python
month = np.arange(2, 13)
plt.plot(month, monthly_growth_rate, marker='o', color='teal')
plt.title("Revenue Growth Rate (%) Trend")
plt.xlabel("Month")
plt.ylabel("Growth Rate (%)")
plt.grid(True)
plt.show()
```

```python
# 3. Cumulative revenue
cumulative_revenue = np.cumsum(monthly_revenue)

for month in range(len(cumulative_revenue)):
    print(f"{month+1} Month's Cumulative Revenue: {cumulative_revenue[month]}")
```

```python
plt.plot(kababjees_sales_analysis[:, 0], cumulative_revenue, marker='o', color='teal')
plt.title("Cumulative Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Cumulative Revenue (PKR)")
plt.grid(True)
plt.show()
```

```python
# 4. Average revenue
avg_revenue = np.mean(monthly_revenue)
print(f"Average revenue of 2024: {avg_revenue}")
```

```python
months = np.arange(1,13)

plt.bar(months, monthly_revenue, color='skyblue', label='Monthly Revenue')
plt.axhline(avg_revenue, color='red', linestyle='--', linewidth=2, label='Average Revenue')

plt.title("Average vs Monthly Revenue (2024)")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.legend(fontsize=8, loc='upper center') 
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Category-Level Analysis

```python
# 1. Total sales of each category -- 2024
categories = ["Burgers", "Pizzas", "Drinks", "Desserts", "Salads"]
category_total = np.sum(monthly_sales, axis=0)

print("Total sales of each category -- Year 2024\n")
for total in range(len(category_total)):
    print(f"{total+1} {categories[total]}: {category_total[total]}")
```

```python
# 2. Which category contributes most to total sales?
max_category = categories[np.argmax(category_total)]
print(f"Top performing category: {max_category}")
```

```python
# 3. Calculate category percentage share
total_revenue = np.sum(category_total)
category_percentage = (category_total/total_revenue)*100
for cat, perc in zip(categories, category_percentage):
    print(f"{cat}: {perc:.2f}%")
```
