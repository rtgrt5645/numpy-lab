# Plotting Graphs Using NumPy

```python
import numpy as np
```

```python
x = np.linspace(-40, 40, 100)
x
```

```python
y = np.sin(x)
y
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot(x,y)
plt.show()
```

```python
y = x*x+2*x+6
```

```python
plt.plot(x, y)
plt.show()
```
